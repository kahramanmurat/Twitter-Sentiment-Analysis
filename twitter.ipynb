# Twitter Text Mining

```python
df = hashtag_df(tweets)

#lower-uppercase transformation
df['text'] = df['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))

#symbols
df['text'] = df['text'].str.replace('[^\w\s]','')

#numbers
df['text'] = df['text'].str.replace('\d','')

#stopwords
import nltk
#nltk.download('stopwords')
from nltk.corpus import stopwords
sw = stopwords.words('english')
df['text'] = df['text'].apply(lambda x: " ".join(x for x in x.split() if x not in sw))

#lemmi
from textblob import Word
#nltk.download('wordnet')
df['text'] = df['text'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()])) 

#nsymbols
df['text'] = df['text'].str.replace('rt','')
```

```python
df["text"]
```

```python
freq_df = df["text"].apply(lambda x: pd.value_counts(x.split(" "))).sum(axis =0).reset_index()
```

```python
freq_df.columns = ["words","frequency"]
```

```python
from textblob import TextBlob
```

```python
def sentiment_skorla(df):

    text = df["text"]

    for i in range(0,len(text)):
        textB = TextBlob(text[i])
        sentiment_skoru = textB.sentiment.polarity
        df.set_value(i, 'sentiment_score', sentiment_skoru)
        
        if sentiment_skoru <0.00:
            duygu_sinifi = 'Negative'
            df.set_value(i, 'emotion_class', duygu_sinifi )

        elif sentiment_skoru >0.00:
            duygu_sinifi = 'Positive'
            df.set_value(i, 'emotion_class', duygu_sinifi )

        else:
            duygu_sinifi = 'Notr'
            df.set_value(i, 'emotion_class', duygu_sinifi )
            
    return df 
```
